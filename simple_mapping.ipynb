```python
# much from http://introtopython.org/visualization_earthquakes.html
# requires an extra package to be installed: basemap
# > conda install basemap
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.basemap import Basemap
```

```python
# make sure the value of resolution is a lowercase L,
#  for 'low', not a numeral 1
plt.figure(figsize=(12,12))
my_map = Basemap(projection='ortho', 
                 lat_0=50, lon_0=-100,
                 resolution='l', area_thresh=1000.0)
my_map.drawcoastlines()
my_map.drawcountries()
my_map.fillcontinents(color='coral')
my_map.drawmapboundary()
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
plt.show()
```

```python
plt.figure(figsize=(16,12))
my_map = Basemap(projection='robin', lat_0=0, lon_0=-100,
              resolution='l', area_thresh=1000.0)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.drawstates()
my_map.fillcontinents(color='skyblue')
my_map.drawmapboundary()
 
my_map.drawmeridians(np.arange(0, 360, 30))
my_map.drawparallels(np.arange(-90, 90, 30))
 
plt.show()
```

```python
plt.figure(figsize=(12,12))
#center=[-122.676207,45.523452] # Portland
center=[-120.0,45.0] 
delta=5.0
my_map = Basemap(projection='merc', lat_0=57, lon_0=-135,
    resolution = 'h', area_thresh = 0.1,
    llcrnrlon=center[0]-delta, llcrnrlat=center[1]-delta,
    urcrnrlon=center[0]+delta, urcrnrlat=center[1]+delta)
 
my_map.drawcoastlines()
my_map.drawcountries()
my_map.drawstates()
#my_map.drawcounties() 
# UnicodeDecodeError: 'utf-8' codec can't decode byte 0xf1 in position 2: invalid continuation byte
my_map.fillcontinents(color='sage')
my_map.drawmapboundary()
 
#my_map.drawmeridians(np.arange(0, 360, 30))
#my_map.drawparallels(np.arange(-90, 90, 30))

lons = [-122.676207]
lats = [45.523452]
x,y = my_map(lons, lats)
my_map.plot(x, y, 'bo', markersize=10)
labels = ['Portland']
for label, xpt, ypt in zip(labels, x, y):
    plt.text(xpt+20000, ypt-20000, label)  # 20 km to the east and 20 km to the south 

plt.show()
```
